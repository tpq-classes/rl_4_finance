### Tossing a Biased Coin

```python
import numpy as np
from numpy.random import default_rng
rng = default_rng(seed=100)
```

```python
ssp = [1, 0]  # <1>
```

```python
asp = [1, 0]  # <2>
```

```python
def epoch():
    tr = 0
    for _ in range(100):
        a = rng.choice(asp)  # <3>
        s = rng.choice(ssp)  # <4>
        if a == s:
            tr += 1  # <5>
    return tr
```

```python
rl = np.array([epoch() for _ in range(250)])  # <6>
rl[:10]
```

```python
rl.mean()  # <7>
```

```python
asp = [1, 0]  # <2>
```

```python
rl.mean()
```

```python
ssp = [1, 1, 1, 1, 0]
```

```python
def epoch(n):
    tr = 0
    asp = [0, 1]  # <1>
    for _ in range(n):
        a = rng.choice(asp)
        s = rng.choice(ssp)
        if a == s:
            tr += 1
        asp.append(s)  # <2>
    return tr
```

```python
rl = np.array([epoch(100) for _ in range(250)])
rl[:10]
```

```python
rl.mean()
```

```python
from collections import Counter
```

```python
ssp = [1, 1, 1, 1, 0]
```

```python
def epoch(n):
    tr = 0
    asp = [0, 1]  # <1>
    for _ in range(n):
        c = Counter(asp)  # <2>
        a = c.most_common()[0][0]  # <3>
        s = rng.choice(ssp)
        if a == s:
            tr += 1
        asp.append(s)  # <4>
    return tr
```

```python
rl = np.array([epoch(100) for _ in range(250)])
rl[:10]
```

```python
rl.mean()
```
