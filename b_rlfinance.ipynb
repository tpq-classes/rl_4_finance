## Single Parameter

See https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html.

```python
def f(x):
    return x ** 2
```

```python
x = np.linspace(-2, 2)
```

```python
y = f(x)
```

```python
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y = f(x)$');
```

```python
from scipy.optimize import minimize
```

```python
opt = minimize(f, 5.5)
opt
```

```python
opt['x']
```

```python
opt['fun']
```

```python
def f(x):
    return (x - 3) ** 2 - x + 1 / 2
```

```python
opt = minimize(f, 5.5)
opt
```

```python
def f(x):
    return x ** 3
```

```python
x = np.linspace(-2, 2)
```

```python
y = f(x)
```

```python
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$y = f(x)$');
```

```python
opt = minimize(f, 5.5)
opt
```

```python
bnds = [(-2, 2)]
```

```python
opt = minimize(f, x0=0.5, bounds=bnds)
opt
```

```python
opt = minimize(lambda x: -f(x), x0=0.5, bounds=bnds)  # maximization
opt
```

```python
opt['x']
```

```python
-opt['fun']
```

```python
def f(x):
    return x ** 2
```

```python
cons = [{'type': 'eq', 'fun': lambda x: x + 2}]
```

```python
opt = minimize(f, x0=0.5, constraints=cons)
opt
```

```python
x = np.linspace(-2, 2)
```

```python
y = f(x)
c = x + 2
```

```python
plt.plot(x, y)
plt.plot(x, c, 'r--')
plt.xlabel('$x$')
plt.ylabel('$y = f(x)$');
```

## Multiple Parameters

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-2, 2)
```

```python
y = np.linspace(-2, 2)
```

```python
x, y = np.meshgrid(x, y)
```

```python
x.shape
```

```python
z = x ** 2 + y ** 2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='coolwarm');
```

```python
def f(p):
    x, y = p
    return x ** 2 + y ** 2
```

```python
f((2, 3))
```

```python
minimize(f, x0=(5, 6))
```

```python
bnds = [(-1.5, 1.5), (-2, 2)]
```

```python
opt = minimize(lambda x: -f(x), x0=(5, 6), bounds=bnds)  # maximization with bounds
opt
```

```python
p = opt['x']
p
```

```python
f(p)
```

```python
z = -(x ** 2 + y ** 2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='coolwarm');
```

```python
def f(p):
    x, y = p
    return -(x ** 2 + y ** 2)
```

```python
minimize(lambda x: -f(x), x0=(5, 6))  # maximization
```
